```python
class LSTMEncDec(nn.Module):
    def __init__(self, input_dim, n_units=256, horizon=6):
        super().__init__()
        self.lstm_enc = nn.LSTM(input_dim, n_units, batch_first=True)
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.lstm_dec = nn.LSTMCell(1, n_units)
        self.horizon = horizon
    
    def forward(self, x):
        output, (h, c) = self.lstm_enc(x)
        h = h[-1]
        c = c[-1]
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h, c = self.lstm_dec(x_inp, (h, c))
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class GRUEncDec(nn.Module):
    def __init__(self, input_dim, n_units=256, horizon=6):
        super().__init__()
        self.gru_enc = nn.GRU(input_dim, n_units, batch_first=True)
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
        
    def forward(self, x):
        output, h = self.gru_enc(x)
        h = h[-1]
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class IMVTensorLSTM1(torch.jit.ScriptModule):
    
    __constants__ = ["n_units", "input_dim"]
    def __init__(self, input_dim, output_dim, n_units, init_std=0.02):
        super().__init__()
        self.U_j = nn.Parameter(torch.randn(input_dim, 1, n_units)*init_std)
        self.U_i = nn.Parameter(torch.randn(input_dim, 1, n_units)*init_std)
        self.U_f = nn.Parameter(torch.randn(input_dim, 1, n_units)*init_std)
        self.U_o = nn.Parameter(torch.randn(input_dim, 1, n_units)*init_std)
        self.W_j = nn.Parameter(torch.randn(input_dim, n_units, n_units)*init_std)
        self.W_i = nn.Parameter(torch.randn(input_dim, n_units, n_units)*init_std)
        self.W_f = nn.Parameter(torch.randn(input_dim, n_units, n_units)*init_std)
        self.W_o = nn.Parameter(torch.randn(input_dim, n_units, n_units)*init_std)
        self.b_j = nn.Parameter(torch.randn(input_dim, n_units)*init_std)
        self.b_i = nn.Parameter(torch.randn(input_dim, n_units)*init_std)
        self.b_f = nn.Parameter(torch.randn(input_dim, n_units)*init_std)
        self.b_o = nn.Parameter(torch.randn(input_dim, n_units)*init_std)
        self.F_alpha_n = nn.Parameter(torch.randn(input_dim, n_units, 1)*init_std)
        self.F_alpha_n_b = nn.Parameter(torch.randn(input_dim, 1)*init_std)
        self.F_beta = nn.Linear(2*n_units, 1)
        self.Phi = nn.Linear(2*n_units, output_dim)
        self.n_units = n_units
        self.input_dim = input_dim
    
    @torch.jit.script_method
    def forward(self, x):
        h_tilda_t = torch.zeros(x.shape[0], self.input_dim, self.n_units).cuda()
        c_tilda_t = torch.zeros(x.shape[0], self.input_dim, self.n_units).cuda()
        outputs = torch.jit.annotate(List[Tensor], [])
        for t in range(x.shape[1]):
            # eq 1
            j_tilda_t = torch.tanh(torch.einsum("bij,ijk->bik", h_tilda_t, self.W_j) + \
                                   torch.einsum("bij,jik->bjk", x[:,t,:].unsqueeze(1), self.U_j) + self.b_j)
            # eq 5
            i_tilda_t = torch.sigmoid(torch.einsum("bij,ijk->bik", h_tilda_t, self.W_i) + \
                                torch.einsum("bij,jik->bjk", x[:,t,:].unsqueeze(1), self.U_i) + self.b_i)
            f_tilda_t = torch.sigmoid(torch.einsum("bij,ijk->bik", h_tilda_t, self.W_f) + \
                                torch.einsum("bij,jik->bjk", x[:,t,:].unsqueeze(1), self.U_f) + self.b_f)
            o_tilda_t = torch.sigmoid(torch.einsum("bij,ijk->bik", h_tilda_t, self.W_o) + \
                                torch.einsum("bij,jik->bjk", x[:,t,:].unsqueeze(1), self.U_o) + self.b_o)
            # eq 6
            c_tilda_t = c_tilda_t*f_tilda_t + i_tilda_t*j_tilda_t
            # eq 7
            h_tilda_t = (o_tilda_t*torch.tanh(c_tilda_t))
            outputs += [h_tilda_t]
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2, 3)
        # eq 8
        alphas = torch.tanh(torch.einsum("btij,ijk->btik", outputs, self.F_alpha_n) +self.F_alpha_n_b)
        alphas = torch.exp(alphas)
        alphas = alphas/torch.sum(alphas, dim=1, keepdim=True)
        g_n = torch.sum(alphas*outputs, dim=1)
        hg = torch.cat([g_n, h_tilda_t], dim=2)
        mu = self.Phi(hg)
        betas = torch.tanh(self.F_beta(hg))
        betas = torch.exp(betas)
        betas = betas/torch.sum(betas, dim=1, keepdim=True)
        mean = torch.sum(betas*mu, dim=1)
        
        return mean
```

```python
from torch.nn.utils import weight_norm
class TCNTemporalBlock(nn.Module):
    def __init__(self, input_channels, output_channels, kernel_size, dilation):
        super().__init__()
        padding = int(dilation*(kernel_size-1))
        self.pad = nn.ConstantPad1d((padding, 0), 0)
        self.conv1 = weight_norm(nn.Conv1d(input_channels, output_channels, kernel_size, dilation=dilation))
        self.conv2 = weight_norm(nn.Conv1d(output_channels, output_channels, kernel_size, dilation=dilation))
        self.residual = nn.Conv1d(input_channels, output_channels, 1)
        
    def forward(self, x):
        out = self.pad(x)
        out = torch.relu(self.conv1(out))
        out = self.pad(out)
        out = torch.relu(self.conv2(out))
        y = self.residual(x)
        out = torch.relu(out + y)
        return out
    

class TCN1(nn.Module):
    def __init__(self, input_dim, n_layers, n_channels=32, kernel_size=3, horizon=6):
        super().__init__()
        self.first_layer = TCNTemporalBlock(input_dim, n_channels, kernel_size, 1)
        self.tcn_layers = nn.ModuleList([TCNTemporalBlock(n_channels, n_channels, kernel_size, 2**(i+1)) for i in range(n_layers-1)])
        self.n_layers = n_layers
        self.output_transform = nn.Linear(n_channels, horizon)
    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.first_layer(x)
        for i in range(self.n_layers-1):
            x = self.tcn_layers[i](x)
        x = x[..., -1]
        out = self.output_transform(x)
        return out
    
class TCN2(nn.Module):
    def __init__(self, input_dim, n_layers, n_channels=32, kernel_size=3, horizon=6):
        super().__init__()
        self.first_layer = TCNTemporalBlock(input_dim, n_channels, kernel_size, 1)
        self.tcn_layers = nn.ModuleList([TCNTemporalBlock(n_channels, n_channels, kernel_size, 2**(i+1)) for i in range(n_layers-1)])
        self.n_layers = n_layers
        self.output_transform = nn.Linear(n_channels, n_channels)
        n_units = n_channels
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
        
    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.first_layer(x)
        for i in range(self.n_layers-1):
            x = self.tcn_layers[i](x)
        x = x[..., -1]
        out = self.output_transform(x)
        
        h = out
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class SE_Block(nn.Module):
    def __init__(self, c, r=16):
        super().__init__()
        self.squeeze = nn.AdaptiveAvgPool1d(1)
        self.excitation = nn.Sequential(
            nn.Linear(c, c // r, bias=False),
            nn.LeakyReLU(inplace=True),
            nn.Linear(c // r, c, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        bs, c, _ = x.shape
        y = self.squeeze(x).view(bs, c)
        y = self.excitation(y).view(bs, c, 1)
        return x * y.expand_as(x)
```

```python
from torch.nn.utils import weight_norm
class TCNTemporalBlock(nn.Module):
    def __init__(self, input_channels, output_channels, kernel_size, dilation):
        super().__init__()
        padding = int(dilation*(kernel_size-1))
        self.pad = nn.ConstantPad1d((padding, 0), 0)
        self.conv1 = weight_norm(nn.Conv1d(input_channels, output_channels, kernel_size, dilation=dilation))
        self.conv2 = weight_norm(nn.Conv1d(output_channels, output_channels, kernel_size, dilation=dilation))
        self.residual = nn.Conv1d(input_channels, output_channels, 1)
        self.se = SE_Block(output_channels)
        
    def forward(self, x):
        out = self.pad(x)
        out = torch.relu(self.conv1(out))
        out = self.pad(out)
        out = torch.relu(self.conv2(out))
        out = self.se(out)
        y = self.residual(x)
        out = torch.relu(out + y)
        return out
    

class TCN3(nn.Module):
    def __init__(self, input_dim, n_layers, n_channels=32, kernel_size=3, horizon=6):
        super().__init__()
        self.first_layer = TCNTemporalBlock(input_dim, n_channels, kernel_size, 1)
        self.tcn_layers = nn.ModuleList([TCNTemporalBlock(n_channels, n_channels, kernel_size, 2**(i+1)) for i in range(n_layers-1)])
        self.n_layers = n_layers
        self.output_transform = nn.Linear(n_channels, horizon)
    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.first_layer(x)
        for i in range(self.n_layers-1):
            x = self.tcn_layers[i](x)
        x = x[..., -1]
        out = self.output_transform(x)
        return out
    
class TCN4(nn.Module):
    def __init__(self, input_dim, n_layers, n_channels=32, kernel_size=3, horizon=6):
        super().__init__()
        self.first_layer = TCNTemporalBlock(input_dim, n_channels, kernel_size, 1)
        self.tcn_layers = nn.ModuleList([TCNTemporalBlock(n_channels, n_channels, kernel_size, 2**(i+1)) for i in range(n_layers-1)])
        self.n_layers = n_layers
        self.output_transform = nn.Linear(n_channels, n_channels)
        n_units = n_channels
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
        
    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.first_layer(x)
        for i in range(self.n_layers-1):
            x = self.tcn_layers[i](x)
        x = x[..., -1]
        out = self.output_transform(x)
        
        h = out
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class ConvBlock(nn.Module):
    def __init__(self, input_dim, kernel_size, n_channels, dropout=0.1):
        super().__init__()
        self.padding = nn.ConstantPad1d(((kernel_size-1)//2, kernel_size//2), 0)
        self.conv = nn.Conv1d(input_dim, n_channels, kernel_size)
        self.batchnorm = nn.BatchNorm1d(n_channels)
        self.dropout = nn.Dropout(dropout)
    def forward(self, x):
        x = self.padding(x)
        x = self.conv(x)
        x = self.batchnorm(x)
        x = self.dropout(x)
        return x

class OSCNN1(nn.Module):
    @staticmethod
    def _get_prime_numbers(N):
        numbers = []
        
        for k in range(1, N + 1):
            condition = True
            for n in range(2, k):
                if k % n == 0:
                    condition = False
                    break
            if condition:
                numbers.append(k)
        return numbers
    def __init__(self, input_dim, seq_length, n_channels=32, dropout=0.1, horizon=6):
        super().__init__()
        
        N = seq_length//4
        
        kernel_sizes = self._get_prime_numbers(N)
        
        self.convs1 = nn.ModuleList([ConvBlock(input_dim, kernel_sizes[i], n_channels, dropout=dropout) for i in range(len(kernel_sizes))])
        
        out_channels1 = n_channels*len(kernel_sizes)
        
        self.convs2 = nn.ModuleList([ConvBlock(out_channels1, kernel_sizes[i], out_channels1//5, dropout=dropout) for i in range(len(kernel_sizes))])
        
        out_channels2 = (out_channels1//5)*len(kernel_sizes)
        
        self.convs31 = ConvBlock(out_channels2, 1, out_channels2)
        self.convs32 = ConvBlock(out_channels2, 2, out_channels2)
        
        self.avg_pooling = nn.AdaptiveAvgPool1d(1)
        
        self.dense = nn.Linear(out_channels2*2, horizon)
        
        self.n_blocks = len(kernel_sizes)
        
    def forward(self, x):
        x = x.permute(0, 2, 1)
        
        first_layer_outputs = []
        for i in range(self.n_blocks):
            out = self.convs1[i](x)
            first_layer_outputs.append(out)
        
        first_layer_outputs = torch.cat(first_layer_outputs, dim=1)
        first_layer_outputs = torch.relu(first_layer_outputs)
        
        second_layer_outputs = []
        for i in range(self.n_blocks):
            out = self.convs2[i](first_layer_outputs)
            second_layer_outputs.append(out)
        
        second_layer_outputs = torch.cat(second_layer_outputs, dim=1)
        second_layer_outputs = torch.relu(second_layer_outputs)
        
        third_layer_output1 = self.convs31(second_layer_outputs)
        third_layer_output2 = self.convs32(second_layer_outputs)
        
        third_layer_outputs = torch.cat([third_layer_output1, third_layer_output2], dim=1)
        third_layer_outputs = torch.relu(third_layer_outputs)
        #print(third_layer_outputs.shape)
        
        out = self.avg_pooling(third_layer_outputs)
        
        out = out.squeeze(2)
        
        out = self.dense(out)
        
        return out
```

```python
class ConvBlock(nn.Module):
    def __init__(self, input_dim, kernel_size, n_channels, dropout=0.1):
        super().__init__()
        self.padding = nn.ConstantPad1d(((kernel_size-1)//2, kernel_size//2), 0)
        self.conv = nn.Conv1d(input_dim, n_channels, kernel_size)
        self.batchnorm = nn.BatchNorm1d(n_channels)
        self.dropout = nn.Dropout(dropout)
    def forward(self, x):
        x = self.padding(x)
        x = self.conv(x)
        x = self.batchnorm(x)
        x = self.dropout(x)
        return x

class OSCNN2(nn.Module):
    @staticmethod
    def _get_prime_numbers(N):
        numbers = []
        
        for k in range(1, N + 1):
            condition = True
            for n in range(2, k):
                if k % n == 0:
                    condition = False
                    break
            if condition:
                numbers.append(k)
        return numbers
    def __init__(self, input_dim, seq_length, n_channels=32, dropout=0.1, horizon=6):
        super().__init__()
        
        N = seq_length//4
        
        kernel_sizes = self._get_prime_numbers(N)
        
        self.convs1 = nn.ModuleList([ConvBlock(input_dim, kernel_sizes[i], n_channels, dropout=dropout) for i in range(len(kernel_sizes))])
        
        out_channels1 = n_channels*len(kernel_sizes)
        
        self.convs2 = nn.ModuleList([ConvBlock(out_channels1, kernel_sizes[i], out_channels1//5, dropout=dropout) for i in range(len(kernel_sizes))])
        
        out_channels2 = (out_channels1//5)*len(kernel_sizes)
        
        self.convs31 = ConvBlock(out_channels2, 1, out_channels2)
        self.convs32 = ConvBlock(out_channels2, 2, out_channels2)
        
        self.avg_pooling = nn.AdaptiveAvgPool1d(1)
        
        self.dense = nn.Linear(out_channels2*2, out_channels2)
        
        self.n_blocks = len(kernel_sizes)
        
        n_units = out_channels2
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
        
    def forward(self, x):
        x = x.permute(0, 2, 1)
        
        first_layer_outputs = []
        for i in range(self.n_blocks):
            out = self.convs1[i](x)
            first_layer_outputs.append(out)
        
        first_layer_outputs = torch.cat(first_layer_outputs, dim=1)
        first_layer_outputs = torch.relu(first_layer_outputs)
        
        second_layer_outputs = []
        for i in range(self.n_blocks):
            out = self.convs2[i](first_layer_outputs)
            second_layer_outputs.append(out)
        
        second_layer_outputs = torch.cat(second_layer_outputs, dim=1)
        second_layer_outputs = torch.relu(second_layer_outputs)
        
        third_layer_output1 = self.convs31(second_layer_outputs)
        third_layer_output2 = self.convs32(second_layer_outputs)
        
        third_layer_outputs = torch.cat([third_layer_output1, third_layer_output2], dim=1)
        third_layer_outputs = torch.relu(third_layer_outputs)
        #print(third_layer_outputs.shape)
        
        out = self.avg_pooling(third_layer_outputs)
        
        out = out.squeeze(2)
        
        out = self.dense(out)
        
        h = out
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class HSGLayer(nn.Module):
    def __init__(self, n_units, init_gates_closed):
        super(HSGLayer, self).__init__()
        self.W_R = nn.Linear(n_units, n_units, bias=False)
        self.W_F = nn.Linear(n_units, n_units)
        if init_gates_closed:
            self.W_F.bias = nn.Parameter(torch.Tensor([-2.5]*n_units).cuda())
    def forward(self, s_L_t, s_prime_tm1):
        g = torch.sigmoid(self.W_R(s_prime_tm1) + self.W_F(s_L_t))
        s_prime_t = g*s_prime_tm1 + (1 - g)*s_L_t
        return s_prime_t

class RHNCell(nn.Module):
    def __init__(self, in_feats, n_units, rec_depth=3, couple_gates=True,
                 use_HSG=False, init_gates_closed=False):
        super(RHNCell, self).__init__()
        self.rec_depth = rec_depth
        self.in_feats = in_feats
        self.n_units = n_units
        self.couple_gates = couple_gates
        self.use_HSG = use_HSG
        self.W_H = nn.Linear(in_feats, n_units, bias=False)
        self.W_T = nn.Linear(in_feats, n_units, bias=False)
        if not couple_gates:
            self.W_C = nn.Linear(in_feats, n_units, bias=False)
        self.R_H = nn.ModuleList([nn.Linear(n_units, n_units) for _ in range(rec_depth)])
        self.R_T = nn.ModuleList([nn.Linear(n_units, n_units) for _ in range(rec_depth)])
        if not couple_gates:
            self.R_C = nn.ModuleList([nn.Linear(n_units, n_units) for _ in range(rec_depth)])
        
        if use_HSG:
            self.HSG = HSGLayer(n_units, init_gates_closed)
        
        if init_gates_closed:
            for l in range(rec_depth):
                self.R_T[l].bias = nn.Parameter(torch.Tensor([-2.5]*n_units).cuda())
                if not couple_gates:
                    self.R_C[l].bias = nn.Parameter(torch.Tensor([-2.5]*n_units).cuda())
        
    def forward(self, x, s):
        if self.use_HSG:
            s_prime_tm1 = s
        preds = []
        for l in range(self.rec_depth):
            if l == 0:
                h_l_t = torch.tanh(self.W_H(x) + self.R_H[l](s))
                t_l_t = torch.sigmoid(self.W_T(x) + self.R_T[l](s))
                if not self.couple_gates:
                    c_l_t = torch.sigmoid(self.W_C(x) + self.R_C[l](s))
            else:
                h_l_t = torch.tanh(self.R_H[l](s))
                t_l_t = torch.sigmoid(self.R_T[l](s))
                if not self.couple_gates:
                    c_l_t = torch.sigmoid(self.R_C[l](s))
            
            if not self.couple_gates:
                s = h_l_t*t_l_t + c_l_t*s
            else:
                s = h_l_t*t_l_t + (1 - t_l_t)*s
            preds.append(s)
                
        if self.use_HSG:
            s = self.HSG(s, s_prime_tm1)
            preds.pop()
            preds.append(s)
        preds = torch.stack(preds)
        return s, preds

    
class RHN(nn.Module):
    def __init__(self, in_feats, n_units=32, rec_depth=3, couple_gates=True, use_HSG=False,
                 init_gates_closed=False, use_batch_norm=False):
        super(RHN, self).__init__()
        assert rec_depth > 0
        self.rec_depth = rec_depth
        self.in_feats = in_feats
        self.n_units = n_units
        self.init_gates_closed = init_gates_closed
        self.couple_gates = couple_gates
        self.use_HSG = use_HSG
        self.use_batch_norm = use_batch_norm
        self.RHNCell = RHNCell(in_feats, n_units, rec_depth, couple_gates=couple_gates,
                               use_HSG=use_HSG, init_gates_closed=init_gates_closed)
        if use_batch_norm:
            self.bn_x = nn.BatchNorm1d(in_feats)
            self.bn_s = nn.BatchNorm1d(n_units)
    def forward(self, x):
        s = torch.zeros(x.shape[0], self.n_units).cuda()
        preds = []
        highway_states = []
        for t in range(x.shape[1]):
            if self.use_batch_norm:
                x_inp = self.bn_x(x[:, t, :])
                s = self.bn_s(s)
            else:
                x_inp = x[:, t, :]
            s, all_s = self.RHNCell(x_inp, s)
            preds.append(s)
            highway_states.append(all_s)
        preds = torch.stack(preds)
        preds = preds.permute(1, 0, 2)
        highway_states = torch.stack(highway_states)
        highway_states = highway_states.permute(2, 0, 3, 1)
        out = preds
        
        return out, highway_states
    
class RHNEncDec(nn.Module):
    def __init__(self, in_feats, n_units, horizon=6):
        super().__init__()
        self.RHN_encoder = RHN(in_feats, n_units)
        self.RHN_decoder = RHNCell(1, n_units)
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(n_units*2, 1)
        self.lrelu = nn.LeakyReLU()
        self.horizon = horizon
    
    def forward(self, x):
        output, _ = self.RHN_encoder(x)
        output = output[:, -1, :]
        s = output
        x_inp = self.output2(self.lrelu(self.output1(s)))
        outputs = []
        for t in range(self.horizon):
            s, _ = self.RHN_decoder(x_inp, s)
            x_inp = self.output2(self.lrelu(self.output1(s)))
            outputs.append(x_inp)
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        return outputs.squeeze(2)
```

```python
class BasicBlock(nn.Module):
    def __init__(self, input_dim, target_dim, n_units=64, hidden_dim=32):
        super().__init__()
        self.FC = nn.Sequential(
            nn.Linear(input_dim, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU())
        
        self.backcast_theta = nn.Linear(n_units, hidden_dim)
        self.forecast_theta = nn.Linear(n_units, hidden_dim)
        
        self.g_b = nn.Linear(hidden_dim, input_dim)
        self.g_f = nn.Linear(hidden_dim, target_dim)
        
    def forward(self, x):
        out = self.FC(x)
        x_hat = self.g_b(self.backcast_theta(out))
        y_hat = self.g_f(self.forecast_theta(out))
        
        return x_hat, y_hat
    
class TrendBlock(nn.Module):
    def __init__(self, input_dim, target_dim, n_units=64, p=3):
        super().__init__()
        self.FC = nn.Sequential(
            nn.Linear(input_dim, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU())
        
        self.backcast_theta = nn.Linear(n_units, p+1)
        self.forecast_theta = nn.Linear(n_units, p+1)
        
        t1 = torch.linspace(0, 1, input_dim).view(-1, 1).cuda()
        t2 = torch.linspace(0, 1, target_dim).view(-1, 1).cuda()
        
        self.T1 = [torch.ones(input_dim).view(-1, 1).cuda()]
        self.T2 = [torch.ones(target_dim).view(-1, 1).cuda()]
        for i in range(1, p+1):
            self.T1.append(t1**i)
            self.T2.append(t2**i)
        self.T1 = torch.cat(self.T1, dim=1)
        self.T2 = torch.cat(self.T2, dim=1)
        
    def forward(self, x):
        out = self.FC(x)
        theta_b = self.backcast_theta(out)
        theta_f = self.forecast_theta(out)
        #print(theta_b, self.T1)
        x_hat = torch.mm(theta_b, self.T1.T)
        y_hat = torch.mm(theta_f, self.T2.T)
        
        return x_hat, y_hat

class SeasonalityBlock(nn.Module):
    def __init__(self, input_dim, target_dim, n_units=64):
        super().__init__()
        self.FC = nn.Sequential(
            nn.Linear(input_dim, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU(),
            nn.Linear(n_units, n_units),
            nn.BatchNorm1d(n_units),
            nn.LeakyReLU())
        
        self.backcast_theta = nn.Linear(n_units, int(input_dim/2-1)*2 + 1)
        self.forecast_theta = nn.Linear(n_units, int(input_dim/2-1)*2 + 1)
        
        t1 = torch.linspace(0, 1, input_dim).view(-1, 1).cuda()
        t2 = torch.linspace(0, 1, target_dim).view(-1, 1).cuda()
        
        self.S1 = [torch.ones(input_dim).view(-1, 1).cuda()]
        self.S2 = [torch.ones(target_dim).view(-1, 1).cuda()]
        for i in range(1, int(input_dim/2-1)+1):
            self.S1.append(torch.cos(2*3.14159*i*t1))
            self.S2.append(torch.cos(2*3.14159*i*t2))
            
        for i in range(1, int(input_dim/2-1)+1):
            self.S1.append(torch.sin(2*3.14159*i*t1))
            self.S2.append(torch.sin(2*3.14159*i*t2))
            
            
        self.S1 = torch.cat(self.S1, dim=1)
        self.S2 = torch.cat(self.S2, dim=1)
        
    def forward(self, x):
        out = self.FC(x)
        theta_b = self.backcast_theta(out)
        theta_f = self.forecast_theta(out)
        x_hat = torch.mm(theta_b, self.S1.T)
        y_hat = torch.mm(theta_f, self.S2.T)
        
        return x_hat, y_hat

class Stack(nn.Module):
    def __init__(self, input_dim, target_dim, n_units=64, block_type='basic', n_blocks=3):
        super().__init__()
        assert block_type in ['basic', 'trend', 'season']
        if block_type == 'basic':
            self.blocks = nn.ModuleList([BasicBlock(input_dim, target_dim, n_units=n_units) for _ in range(n_blocks)])
        elif block_type == 'trend':
            self.blocks = nn.ModuleList([TrendBlock(input_dim, target_dim, n_units=n_units) for _ in range(n_blocks)])
        elif block_type == 'season':
            self.blocks = nn.ModuleList([SeasonalityBlock(input_dim, target_dim, n_units=n_units) for _ in range(n_blocks)])
        
        self.n_blocks = n_blocks
        self.target_dim = target_dim
    def forward(self, x):
        y_hat_sum = torch.zeros(x.shape[0], self.target_dim).cuda()
        for i in range(self.n_blocks):
            x_hat, y_hat = self.blocks[i](x)
            x = x - x_hat
            y_hat_sum += y_hat
        return x, y_hat_sum
    
    
class NBeats(nn.Module):
    def __init__(self, input_dim, target_dim, n_units=64, blocks_in_stack=3,
                 n_trend_stacks=1, n_season_stacks=1, n_basic_stacks=5):
        super().__init__()
        self.trend_stacks = nn.ModuleList([Stack(input_dim, target_dim, n_units=n_units, block_type='trend',
                                                n_blocks=blocks_in_stack) for _ in range(n_trend_stacks)])
        self.season_stacks = nn.ModuleList([Stack(input_dim, target_dim, n_units=n_units, block_type='season',
                                                n_blocks=blocks_in_stack) for _ in range(n_season_stacks)])
        self.basic_stacks = nn.ModuleList([Stack(input_dim, target_dim, n_units=n_units, block_type='basic',
                                                n_blocks=blocks_in_stack) for _ in range(n_basic_stacks)])
        
        self.n_trend_stacks = n_trend_stacks
        self.n_season_stacks = n_season_stacks
        self.n_basic_stacks = n_basic_stacks
        self.target_dim = target_dim
    
    def forward(self, x):
        y_hat_sum = torch.zeros(x.shape[0], self.target_dim).cuda()
        for i in range(self.n_trend_stacks):
            x, y_hat = self.trend_stacks[i](x)
            y_hat_sum += y_hat
        
        for i in range(self.n_season_stacks):
            x, y_hat = self.season_stacks[i](x)
            y_hat_sum += y_hat
        
        for i in range(self.n_basic_stacks):
            x, y_hat = self.basic_stacks[i](x)
            y_hat_sum += y_hat
        
        return y_hat_sum
    
    
class MVNBeats1(nn.Module):
    def __init__(self, input_dim, seq_length, n_units=64, blocks_in_stack=3,
                 n_trend_stacks=1, n_season_stacks=1, n_basic_stacks=5, horizon=6):
        super().__init__()
        self.inp_transform1 = nn.Linear(input_dim, 2*input_dim)
        self.inp_transform2 = nn.Linear(input_dim*2, 1)
        self.lrelu1 = nn.LeakyReLU()
        self.nbeats = NBeats(seq_length, n_units//4, n_units, blocks_in_stack, n_trend_stacks,
                                            n_season_stacks, n_basic_stacks)
        self.out1 = nn.Linear(n_units//4, n_units)
        self.out2 = nn.Linear(n_units, horizon)
        self.lrelu = nn.LeakyReLU()
        self.input_dim = input_dim
    
    def forward(self, x):
        x = self.inp_transform2(self.lrelu1(self.inp_transform1(x)))
        x = x.squeeze(2)
        outs = self.nbeats(x)
        out = self.out2(self.lrelu(self.out1(outs)))
        
        return out
    
class MVNBeats2(nn.Module):
    def __init__(self, input_dim, seq_length, n_units=64, blocks_in_stack=3,
                 n_trend_stacks=1, n_season_stacks=1, n_basic_stacks=5, horizon=6):
        super().__init__()
        self.inp_transform1 = nn.Linear(input_dim, 2*input_dim)
        self.inp_transform2 = nn.Linear(input_dim*2, 1)
        self.lrelu1 = nn.LeakyReLU()
        self.nbeats = NBeats(seq_length, n_units//4, n_units, blocks_in_stack, n_trend_stacks,
                                            n_season_stacks, n_basic_stacks)
        self.out1 = nn.Linear(n_units//4, n_units)
        self.out2 = nn.Linear(n_units, n_units)
        self.lrelu = nn.LeakyReLU()
        self.input_dim = input_dim
        
        
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
    
    def forward(self, x):
        x = self.inp_transform2(self.lrelu1(self.inp_transform1(x)))
        x = x.squeeze(2)
        outs = self.nbeats(x)
        out = self.out2(self.lrelu(self.out1(outs)))
        
        h = out
        x_inp = self.output2(self.lrelu(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class SpatialGatingUnit(nn.Module):
    def __init__(self, d_ffn, seq_len):
        super().__init__()

        self.norm = nn.LayerNorm(d_ffn // 2)
        self.proj = nn.Linear(seq_len, seq_len)

    def forward(self, x):
        u, v = x.chunk(2, dim=-1)
        v = self.norm(v)
        v = v.permute(0, 2, 1)
        v = self.proj(v)
        v = v.permute(0, 2, 1)
        return u * v


class GatingMlpBlock(nn.Module):
    def __init__(self, d_model, d_ffn, seq_len, survival_prob):
        super().__init__()

        self.norm = nn.LayerNorm(d_model)
        self.proj_1 = nn.Linear(d_model, d_ffn)
        self.activation = nn.GELU()
        self.spatial_gating_unit = SpatialGatingUnit(d_ffn, seq_len)
        self.proj_2 = nn.Linear(d_ffn // 2, d_model)
        self.prob = survival_prob
        self.m = torch.distributions.bernoulli.Bernoulli(torch.Tensor([self.prob]))

    def forward(self, x):
        if self.training and torch.equal(self.m.sample(), torch.zeros(1)):
            return x
        shorcut = x.clone()
        x = self.norm(x)
        x = self.proj_1(x)
        x = self.activation(x)
        x = self.spatial_gating_unit(x)
        x = self.proj_2(x)
        return x + shorcut


class gMLP1(nn.Module):
    def __init__(
        self,
        d_model,
        d_ffn,
        seq_len,
        n_blocks,
        prob_0_L=[1, 0.5],
        horizon=6
    ):
        super().__init__()
        self.inp_transform1 = nn.Linear(d_model, 2*d_model)
        self.lrelu = nn.LeakyReLU()
        self.inp_transform2 = nn.Linear(d_model*2, 1)
        self.survival_probs = torch.linspace(prob_0_L[0], prob_0_L[1], n_blocks)
        self.blocks = nn.ModuleList(
            [GatingMlpBlock(d_model, d_ffn, seq_len, prob) for prob in self.survival_probs]
        )
        self.out_transform1 = nn.Linear(d_model, 2*d_model)
        self.out_transform2 = nn.Linear(2*d_model, 1)
        self.lrelu1 = nn.LeakyReLU()
        self.out_transform3 = nn.Linear(seq_len, horizon)
    def forward(self, x):
        #x = self.inp_transform2(self.lrelu(self.inp_transform1(x)))
        #x = x.squeeze(2)
        for gmlp_block in self.blocks:
            x = gmlp_block(x)
            
        x = self.out_transform2(self.lrelu1(self.out_transform1(x)))
        x = x.squeeze(2)
        x = self.out_transform3(x)
        return x

class gMLP2(nn.Module):
    def __init__(
        self,
        d_model,
        d_ffn,
        seq_len,
        n_blocks,
        prob_0_L=[1, 0.5],
        horizon=6
    ):
        super().__init__()
        self.inp_transform1 = nn.Linear(d_model, 2*d_model)
        self.lrelu = nn.LeakyReLU()
        self.inp_transform2 = nn.Linear(d_model*2, 1)
        self.survival_probs = torch.linspace(prob_0_L[0], prob_0_L[1], n_blocks)
        self.blocks = nn.ModuleList(
            [GatingMlpBlock(d_model, d_ffn, seq_len, prob) for prob in self.survival_probs]
        )
        self.out_transform1 = nn.Linear(d_model, 2*d_model)
        self.out_transform2 = nn.Linear(2*d_model, 1)
        self.lrelu1 = nn.LeakyReLU() 
        n_units = seq_len
        self.output1 = nn.Linear(n_units, 2*n_units)
        self.output2 = nn.Linear(2*n_units, 1)
        self.lrelu2 = nn.LeakyReLU()
        self.gru_dec = nn.GRUCell(1, n_units)
        self.horizon = horizon
    def forward(self, x):
        for gmlp_block in self.blocks:
            x = gmlp_block(x)
            
        x = self.out_transform2(self.lrelu1(self.out_transform1(x)))
        x = x.squeeze(2)
        
        h = x
        x_inp = self.output2(self.lrelu2(self.output1(h)))
        outputs = []
        for t in range(self.horizon):
            h = self.gru_dec(x_inp, h)
            x_inp = self.output2(self.lrelu2(self.output1(h)))
            outputs.append(x_inp)
            
        outputs = torch.stack(outputs)
        outputs = outputs.permute(1, 0, 2)
        
        return outputs.squeeze(2)
```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()       
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        #pe.requires_grad = False
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:x.size(0), :]
          

class TransAm1(nn.Module):
    def __init__(self, input_dim, seq_len, feature_size=250,num_layers=1,dropout=0.1, horizon=6):
        super().__init__()
        self.inp_transform1 = nn.Linear(input_dim, 2*input_dim)
        self.lrelu = nn.LeakyReLU()
        self.inp_transform2 = nn.Linear(2*input_dim, 1)
        self.model_type = 'Transformer'
        self.src_mask = None
        self.pos_encoder = PositionalEncoding(feature_size)
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=feature_size, nhead=10, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=num_layers)        
        self.decoder = nn.Linear(feature_size,1)
        self.decoder1 = nn.Linear(seq_len, horizon)
        self.init_weights()

    def init_weights(self):
        initrange = 0.1    
        self.decoder.bias.data.zero_()
        self.decoder.weight.data.uniform_(-initrange, initrange)

    def forward(self,src):
        src = self.inp_transform2(self.lrelu(self.inp_transform1(src)))
        src = src.permute(1, 0, 2)
        if self.src_mask is None or self.src_mask.size(0) != len(src):
            device = src.device
            mask = self._generate_square_subsequent_mask(len(src)).to(device)
            self.src_mask = mask

        src = self.pos_encoder(src)
        output = self.transformer_encoder(src,self.src_mask)#, self.src_mask)
        output = self.decoder(output)
        output = output.squeeze(2)
        output = output.permute(1, 0)
        output = self.decoder1(output)
        return output

    def _generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
```
